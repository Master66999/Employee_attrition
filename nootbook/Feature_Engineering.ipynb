```python
# ==========================================================
# Employee Attrition Analysis - Phase 6: Feature Engineering
# ==========================================================

import pandas as pd
import numpy as np

# ----------------------------------------------------------
# 1. Load Dataset
# ----------------------------------------------------------

df = pd.read_csv("/content/employee_attrition_processed.csv")

print("Dataset Shape:", df.shape)
print(df.head())

# ----------------------------------------------------------
# 2. Create Age Groups
# ----------------------------------------------------------

df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60']
)

# ----------------------------------------------------------
# 3. Income Per Working Year
# ----------------------------------------------------------

df['IncomePerWorkingYear'] = (
    df['MonthlyIncome'] /
    (df['TotalWorkingYears'] + 1)
)

# ----------------------------------------------------------
# 4. Promotion Delay
# ----------------------------------------------------------

df['PromotionDelay'] = (
    df['YearsAtCompany'] -
    df['YearsSinceLastPromotion']
)

# ----------------------------------------------------------
# 5. Company Experience Ratio
# ----------------------------------------------------------

df['CompanyExperienceRatio'] = (
    df['YearsAtCompany'] /
    (df['TotalWorkingYears'] + 1)
)

# ----------------------------------------------------------
# 6. Early Career Employee Flag
# ----------------------------------------------------------

df['EarlyCareerEmployee'] = (
    df['YearsAtCompany'] < 3
).astype(int)

# ----------------------------------------------------------
# 7. Long Commute Flag
# ----------------------------------------------------------

df['LongCommute'] = (
    df['DistanceFromHome'] > 20
).astype(int)

# ----------------------------------------------------------
# 8. High Income Employee Flag
# ----------------------------------------------------------

salary_threshold = df['MonthlyIncome'].median()

df['HighIncomeEmployee'] = (
    df['MonthlyIncome'] > salary_threshold
).astype(int)

# ----------------------------------------------------------
# 9. Promotion Frequency
# ----------------------------------------------------------

df['PromotionFrequency'] = (
    df['YearsSinceLastPromotion'] /
    (df['YearsAtCompany'] + 1)
)

# ----------------------------------------------------------
# 10. Income Growth Index
# ----------------------------------------------------------

df['IncomeGrowthIndex'] = (
    df['MonthlyIncome'] /
    (df['Age'] + 1)
)

# ----------------------------------------------------------
# 11. Role Stability
# ----------------------------------------------------------

df['RoleStability'] = (
    df['YearsInCurrentRole'] /
    (df['YearsAtCompany'] + 1)
)

# ----------------------------------------------------------
# 12. Manager Stability
# ----------------------------------------------------------

df['ManagerStability'] = (
    df['YearsWithCurrManager'] /
    (df['YearsAtCompany'] + 1)
)

# ----------------------------------------------------------
# 13. Training Intensity
# ----------------------------------------------------------

df['TrainingIntensity'] = (
    df['TrainingTimesLastYear'] /
    (df['TotalWorkingYears'] + 1)
)

# ----------------------------------------------------------
# 14. Encode AgeGroup
# ----------------------------------------------------------

df = pd.get_dummies(
    df,
    columns=['AgeGroup'],
    drop_first=True
)

# Convert boolean columns to integer
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)

# ----------------------------------------------------------
# 15. Correlation with Attrition
# ----------------------------------------------------------

correlation = (
    df.corr(numeric_only=True)['Attrition']
    .sort_values(ascending=False)
)

print("\nTop Features Correlated with Attrition:\n")
print(correlation.head(15))

# ----------------------------------------------------------
# 16. Save Feature Engineered Dataset
# ----------------------------------------------------------

df.to_csv(
    "employee_attrition_feature_engineered.csv",
    index=False
)

print("\nFeature Engineering Completed Successfully!")
print("Final Dataset Shape:", df.shape)
print("Dataset saved as:")
print("employee_attrition_feature_engineered.csv")
```
