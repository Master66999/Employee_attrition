```python
columns_to_drop = [
    'EmployeeCount',
    'EmployeeNumber',
    'Over18',
    'StandardHours'
]

df.drop(columns=columns_to_drop, inplace=True)
```

```python
df['Attrition'].value_counts()
```

```python
df['Attrition'] = df['Attrition'].map(
    {'Yes': 1, 'No': 0}
)
```

```python
df['Attrition'].value_counts()
```

```python
from pandas.core.arrays import categorical
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
df = pd.get_dummies(
    df,
    columns=categorical_columns,
    drop_first=True
)
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
print(x.shape)
print(y.shape)
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

x_balanced, y_balanced = smote.fit_resample(x, y)
```

```python
pd.Series(y_balanced).value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_balanced,
    y_balanced,
    test_size=0.20,
    random_state=42,
    stratify=y_balanced
)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
processed_df = pd.concat(
    [
        pd.DataFrame(x_balanced),
        pd.Series(y_balanced, name='Attrition')
    ],
    axis=1
)
```

```python
processed_df.to_csv(
    "employee_attrition_processed.csv",
    index=False
)
```
