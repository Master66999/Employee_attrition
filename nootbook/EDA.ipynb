```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Attrition'].value_counts()
```

```python
sns.countplot(x='Attrition', data=df)
plt.title ("employee distribution")
plt.show()
```

```python
department_attrition = pd.crosstab(
    df['Department'],
    df['Attrition']
)

department_attrition
```

```python
department_attrition.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5)
)

plt.title('Attrition by Department')
plt.ylabel('Employees')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    x = 'JobRole',
    hue = 'Attrition',
    data = df

)

plt.xticks(rotation=90)
plt.title('Attrition by Job Role')
plt.show()

```

```python
sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df
)

plt.title('Attrition by Over Time')
plt.show()

```

```python
plt.figure(figsize =(10,5))

sns.boxplot(
    x = 'Attrition',
    y = 'MonthlyIncome',
    data = df

)

plt.show()
```

```python
plt.figure(figsize = (10,5))

sns.histplot(
    data = df,
    x = 'Age',
    hue = 'Attrition',
    bins = 20,
    kde = True

)

plt.show()
```

```python
sns.countplot(
    x = 'JobSatisfaction',
    hue = 'Attrition',
    data = df
)

plt.title('Attrition by Job Satisfaction')
plt.show()
```

```python
sns.countplot(
    x = 'WorkLifeBalance',
    hue = 'Attrition',
    data = df
)

plt.title('Attrition by Job Satisfaction')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='Attrition',
    y='DistanceFromHome',
    data=df
)

plt.show()
```

```python
plt.figure(figsize = (10,5))

sns.boxplot(
    x = 'YearsAtCompany',
    y = 'DistanceFromHome',
    data = df
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='Attrition',
    y='YearsAtCompany',
    data=df
)

plt.show()
```

```python
numeric_df = df.select_dtypes(
    include=['int64', 'float64']
)
```

```python
plt.figure(figsize=(18,12))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.show()
```

```python
gender_attrition = pd.crosstab(
    df['Gender'],
    df['Attrition'],
    normalize='index'
) * 100

gender_attrition
```

```python
gender_attrition.plot(
    kind='bar',
    figsize=(6,4)
)

plt.ylabel('Percentage')
plt.show()
```
